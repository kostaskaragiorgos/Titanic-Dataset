```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier,ExtraTreesRegressor , RandomForestRegressor
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
import pickle
from sklearn.neural_network import MLPClassifier
```

```python
df = pd.read_csv('submission files/test.csv')
df.head()
```

```python
print(df.info())
df['Age'] = df['Age'].fillna(df['Age'].std())
df['Age'] = df['Age'].astype('float32')
print(df.info())
labels = ['child', 'young adult','adult', 'old']
df['age_group'] = pd.cut(df['Age'], bins=4, labels=labels)
df['age_group']= df['age_group'].fillna('adult')
df.isnull().sum()
```

```python
labels = ['very cheap','cheap', 'normal', 'expensive']
df['Fare_group'] = pd.cut(df['Fare'], bins=4, labels=labels)
df['Fare_group']= df['Fare_group'].fillna('cheap')
print(df.isnull().sum())
df.info()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df.head()
```

```python
df.loc[df['FamilySize'] == 1, 'IsAlone'] = 1
df['IsAlone']= df['IsAlone'].fillna(0)
df['IsAlone'] = df['IsAlone'].astype('int32')
df
```

```python
df['Has_Cabin'] = df["Cabin"].apply(lambda x: 0 if type(x) == float else 1)
df.head()
```

```python
df['Embarked']= df['Embarked'].fillna('S')
df.info()
```

```python
df.drop(columns= ['PassengerId', 'Name', 'Cabin', 'Ticket','Age', 'Fare', 'SibSp', 'Parch'], inplace=True)
df
```

```python
X = df.iloc[:, :].values
X
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1,2,3,4])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
sc = StandardScaler()
X[:, :] = sc.fit_transform(X[:, :])
print(X)
```

```python
filename = 'LogisticRegression_model_newapp3081.sav'
loaded_modelmlp = pickle.load(open(filename, 'rb'))
```

```python
predictions= loaded_modelmlp.predict(X)
print(predictions)
```

```python
sub = pd.read_csv('submission files/gender_submission.csv', delimiter=',')
sub.drop(columns=['Survived'], inplace=True)
sub['Survived'] = predictions
sub
sub.to_csv('submission files/readylog81.csv', index=False)
```
